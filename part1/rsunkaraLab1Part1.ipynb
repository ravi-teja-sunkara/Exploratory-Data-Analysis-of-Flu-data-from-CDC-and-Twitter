## Problem 1: 
Define two synthetic vectors of data representing sales over 12 months for 2 items. Compare the two using lines graphs. (Discussion; Run the code multiple time and see randomness of the second set of data..)

```r
sales1 <- c(12,14,16,29,30,45,19,20,16, 19, 34, 20)
#rpois(n, lambda) - n is the # of random values to return with a mean of 'lambda'
sales2 <- rpois(12, 34) 
```

```r
sales2
print(sum(sales1))
print(sum(sales2))
```

```r
# If you set parameter values here, the changes will be in effect for the rest of the session or until you change them again. 
par(bg='cornsilk') # bg sets the background

# This plot shows both the sales that is 1 and 2.
plot(sales1, col='blue', type = 'o', ylim = c(0, 100), xlab = "Month", ylab="Sales")
title(main = 'Sales by Month')

# lines() adds information to the graph. It cannot generate a graph on its own
lines(sales2, type = 'o', pch = 22, lty=2, col='red') 
grid(nx=NA, ny=NULL)
legend('topright', inset = .05, c("Sales1", "Sales2"), fill=c("blue", 'red'), horiz = TRUE)
```

### Discussion:
The number of sales per month of 'sales2' is random and it can be more or less than the 'sales1' in a given month.
But, the total number of sales over 12 months for 'sales2' will always be greater than(>) 'sales1'.

## Problem 3: 
Use boxplot to compare the two sales data. (Discussion: How will you interpret the graph visualization?)

```r
par(bg='cornsilk')
fn <- boxplot(sales, col=c("orange", "green"))$stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0.4, cex=.75)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0.1, cex=.75)
grid(nx=NA, ny=NULL)
```

### Discussion:
The median of 'Sales2' is more than the 'Sales1' as can be seen from the boxplot. Both the boxplots of 'Sales1' and 'Sales2' are skewed towards the lower end that is towards right. We can compute the range of the data from the boxplot as shown above 'Sales1' has a range of 68 (80-12= 68) and 'Sales2' has a range of 66 (89-23= 66).
